```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Indian%20Unicorn%20startups%202023%20updated.csv')
print(df)
```

missinga values

```python
print("missing values")
df.isnull().sum()
```

Fill na values if any

```python
print(df.isnull().any(axis=1))
df[df.isnull().any(axis=1)]
```

Fill na values if any

```python
df['No.']=df['No.'].fillna(df['No.'].mean())
df['Valuation ($B)']=df['Valuation ($B)'].fillna(df['Valuation ($B)'].mean())
```

Check no of duplicate rows

```python
df.duplicated().sum()
print(df.duplicated().sum())
print(df.duplicated())
```

Basic statistics of Dataset( use describe method)

```python
df.describe()
```

Clean column names( remove spaces at first and last of column name if present

```python
print("before cleaning")
print(df.columns)
df.columns = df.columns.str.strip()
print("after cleaning")
print(df.columns)
```

Convert Entry Column To DateTime

```python
df['Entry date'] = pd.to_datetime(df['Entry'], format='%b/%Y', errors='coerce')
df.head()
```

Create Valuation_Growth Column ( valuation/entry_valuation)

```python
df['Valuation_Growth'] = df['Valuation ($B)'] / df['Entry Valuation^^ ($B)']
df.head()
```

Create Primary_City Column( First word in location column)  use split function

```python
df['Primary_City'] = df['Location'].str.split('/').str[0]
df.head()
```

Create No of Investors Column( Count in  Select investors column)

```python
df['No of Investors'] = df['Select Investors'].str.split(',').str.len()
df.head()
```

```python
import pandas as pd
url="https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Indian%20Unicorn%20startups%202023%20updated.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
x=df[['Entry Valuation^^ ($B)']].values
y=df[['Valuation ($B)']].values
sc_x=StandardScaler()

sc_y=StandardScaler()
x_scaled=sc_x.fit_transform(x)
y_scaled=sc_y.fit_transform(y)
```

```python
y_scaled
```

```python
y_scaled=y_scaled.ravel()
```

```python
y_scaled
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf',C=100, epsilon=0.1)
svr.fit(x_scaled,y_scaled)
```

```python
y_pred_scaled=svr.predict(x_scaled)
y_pred=sc_y.inverse_transform(y_pred_scaled.reshape(-1,1))
```

```python
y_pred_scaled
```

```python
y_pred
```

```python
dframe=pd.DataFrame({'Actual':y.ravel(),'Predicted':y_pred.ravel()})
dframe.head(15)
```

```python

from sklearn.metrics import mean_squared_error
import numpy as np
mse=mean_squared_error(y,y_pred)
rmse=np.sqrt(mse)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
```

```python
import matplotlib.pyplot as plt
graph=dframe.head(10)
graph.plot(kind='bar')
plt.title('Actual vs Predicted')
plt.xlabel('Sector')
plt.ylabel('No.')
```
